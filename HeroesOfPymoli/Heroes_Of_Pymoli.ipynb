```python
import pandas as pd
```

```python
file = "Resources/purchase_data.csv"
purchase_data = pd.read_csv(file)
```

```python
purchase_data.head()
```

```python
player_count = len(purchase_data["SN"].value_counts())
player_count
```

```python
Number_of_Unique_Items = len(purchase_data["Item Name"].value_counts())
Average_Purchasing_Price = (purchase_data["Price"].mean())
Total_Purchases = (purchase_data["Purchase ID"].count())
Total_Revenue = sum(purchase_data["Price"])
Purchasing_Analysis_Total_df = pd.DataFrame({"Number of Unique Items":[Number_of_Unique_Items],"Average Purchasing Price":[Average_Purchasing_Price],"Total Number of Purchases":[Total_Purchases],"Total Revenue":[Total_Revenue]})
Purchasing_Analysis_Total_df["Average Purchasing Price"] = Purchasing_Analysis_Total_df["Average Purchasing Price"].map("${:.2f}".format)
Purchasing_Analysis_Total_df["Total Revenue"] = Purchasing_Analysis_Total_df["Total Revenue"].map("${:.2f}".format)
Purchasing_Analysis_Total_df
```

```python
Gender_mix = purchase_data.groupby("Gender")
Gender_Count = Gender_mix["SN"].nunique()
Gender_Percentage = (Gender_Count / player_count)*100
Gender_Demographic_df = pd.DataFrame({"Total Count":Gender_Count,"Percentage of Players":Gender_Percentage})
Gender_Demographic_df["Percentage of Players"] = Gender_Demographic_df["Percentage of Players"].map("{:.2f}%".format)
Gender_Demographic_df
```

```python
Purchase_Count = Gender_mix["Purchase ID"].count()
Average_Purchase_Price = Gender_mix["Price"].mean()
Total_Purchase_Value = Gender_mix["Price"].sum()
Average_Purchase_Total_Per_Person = Total_Purchase_Value / Gender_Count
Purchasing_Analysis_Gender_df = pd.DataFrame({"Purchase Count":Purchase_Count,"Average Purchase Price":Average_Purchase_Price,"Total Purchase Value":Total_Purchase_Value,"Average Purchase Total Per Person":Average_Purchase_Total_Per_Person})
Purchasing_Analysis_Gender_df["Average Purchase Price"] = Purchasing_Analysis_Gender_df["Average Purchase Price"].map("${:.2f}".format)
Purchasing_Analysis_Gender_df["Total Purchase Value"] = Purchasing_Analysis_Gender_df["Total Purchase Value"].map("${:.2f}".format)
Purchasing_Analysis_Gender_df["Average Purchase Total Per Person"] = Purchasing_Analysis_Gender_df["Average Purchase Total Per Person"].map("${:.2f}".format)
Purchasing_Analysis_Gender_df
```

```python
Bins = [0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,44.9,99999]
Age_Ranges= ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40-44","45+"]
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], Bins, labels=Age_Ranges)
Age_mix = purchase_data.groupby("Age Group")
Total_Player_Count_by_Age = Age_mix["SN"].nunique()
Percentage_by_Age = (Total_Player_Count / player_count)*100
Age_Demographic_df = pd.DataFrame({"Total Player Count":Total_Player_Count_by_Age,"Percentage of Players":Percentage_by_Age})
Age_Demographic_df["Percentage of Players"] = Age_Demographic_df["Percentage of Players"].map("{:.2f}%".format)
Age_Demographic_df
```

```python
Purchase_Count_Age = Age_mix["Purchase ID"].count()
Average_Purchase_Price_Age = Age_mix["Price"].mean()
Total_Purchase_Value_Age = Age_mix["Price"].sum()
Average_Purchase_Total_Per_Person_Age = Total_Purchase_Value_Age / Total_Player_Count_by_Age
Purchasing_Analysis_Age_df = pd.DataFrame({"Purchase Count":Purchase_Count_Age,"Average Purchase Price":Average_Purchase_Price_Age,"Total Purchase Value":Total_Purchase_Value_Age,"Average Purchase Total Per Person":Average_Purchase_Total_Per_Person_Age})
Purchasing_Analysis_Age_df["Average Purchase Price"] = Purchasing_Analysis_Age_df["Average Purchase Price"].map("${:.2f}".format)
Purchasing_Analysis_Age_df["Total Purchase Value"] = Purchasing_Analysis_Age_df["Total Purchase Value"].map("${:.2f}".format)
Purchasing_Analysis_Age_df["Average Purchase Total Per Person"] = Purchasing_Analysis_Age_df["Average Purchase Total Per Person"].map("${:.2f}".format)
Purchasing_Analysis_Age_df
```

```python
Spender_Mix = purchase_data.groupby("SN")
Spender_Purchase_Count = Spender_Mix["Purchase ID"].count()
Spender_Average_Purchase_Price = Spender_Mix["Price"].mean()
Spender_Total_Purchase_Value = Spender_Mix["Price"].sum()
Top_Spenders_df = pd.DataFrame({"Purchase Count":Spender_Purchase_Count,"Average Purchase Price":Spender_Average_Purchase_Price,"Total Purchase Value":Spender_Total_Purchase_Value})
Top_Spenders_df["Average Purchase Price"] = Top_Spenders_df["Average Purchase Price"].map("${:.2f}".format)
Top_Spenders_df["Total Purchase Value"] = Top_Spenders_df["Total Purchase Value"].map("${:.2f}".format)
Top_Spenders_df.sort_values(["Purchase Count"],ascending=False).head(5)
```

```python
Items = purchase_data[["Item ID","Item Name","Price"]]
Popular_Items = Items.groupby(["Item ID","Item Name"])
Items_Count = Popular_Items["Price"].count()
Total_Purchase_Value = Popular_Items["Price"].sum()
Items_Price = Total_Purchase_Value / Items_Count
Most_Popular_Items_df = pd.DataFrame({"Purchase Count":Items_Count,"Item Price":Items_Price,"Total Purchase Value":Total_Purchase_Value})
Most_Popular_Items_df["Item Price"] = Most_Popular_Items_df["Item Price"].map("${:.2f}".format)
Most_Popular_Items_df["Total Purchase Value"] = Most_Popular_Items_df["Total Purchase Value"].map("${:.2f}".format)
Most_Popular_Items_df.sort_values(["Purchase Count"],ascending=False).head(5)
```
